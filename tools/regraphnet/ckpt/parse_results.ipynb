# Parse Result Log

```python
import os
import json
from pathlib import Path
import pandas as pd
```

```python
input_file = Path("model_v42_synth_results.json")

with open(input_file) as f:
    results = json.load(f)

results_df = pd.DataFrame(results)
results_df["accuracy"] = results_df[["start_acc", "end_acc", "operation_acc"]].sum(axis=1)
results_df
```

```python
train_df = results_df[results_df["train_test"] == "Train"]
train_df_sorted = train_df.sort_values("loss")
train_df_sorted
```

```python
lowest_train_lost_row = train_df_sorted.iloc[0].name
test_df = results_df[results_df["train_test"] == "Test"]
test_df.loc[lowest_train_lost_row + 1]
```

```python
test_df_sorted = test_df.sort_values("accuracy", ascending=False)
test_df_sorted
```
